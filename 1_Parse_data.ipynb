```python
mouse_movement_data_copy=mouse_movement_data.copy()

def convert(data):
    def convert():
        for d in data:
            #print(d)
            for k, v in d.items():
                for d2 in v['data']:
                    d3 = dict(d2)
                    d3.update(v)
                    del d3['data']
                    #print(d3)
                    d3['session_id'] = k
                    #print(d3)
                    
                    yield d3
    return list(convert())


data2 = convert(mouse_movement_data['sessions'])

df = pd.DataFrame(data2)

df.head()
```

```python
def parse_session_id(df):
    new_df=pd.DataFrame(columns=df.columns)
    indexx=1
    for i in range(len(df)):
        array_to_add = df.loc[i]
        for session_id in df.loc[i]['array_session_id']:
            array_to_add['array_session_id']=session_id
            new_df.loc[indexx]=array_to_add
            indexx+=1
    return new_df
```
